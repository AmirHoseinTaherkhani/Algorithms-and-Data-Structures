## Problem 0-1 
Let 
A = {i + nCr(i,5)}, where i ∈ Z and 0 ≤ i ≤ 4 and

B = {3i | i ∈ {1, 2, 4, 5}

Evaluate: (a) A ∩ B

(b) |A ∪ B|

(c) |A − B|

```python
import math
A = set()
B = set()
for i in range(5):
    A.add(i + (math.factorial(5)/(math.factorial(i)*math.factorial(5-i))))
for j in [1, 2, 4, 5]:
    B.add(3*j)
print(f'A ∩ B = {A.intersection(B)}')
print(f'|A ∪ B| = {len(A.union(B))}')
print(f'|A − B| = {len(A.difference(B))}')
```

## Problem 0-2

Let X be the random variable representing the number of heads seen after flipping
a fair coin three times. Let Y be the random variable representing the outcome of rolling two fair
six-sided dice and multiplying their values. Please compute the following expected values.

```python
X = []
for i in range(4):
    for _ in range(int(math.factorial(3)/(math.factorial(i)*math.factorial(3-i)))):
        X.append(i)


Y = []
for i in range(1,7):
    for j in range(1,7):
        Y.append(i*j)

print(f'E[X] = {sum(X)/len(X)}') 
print(f'E[Y] = {sum(Y)/len(Y)}')
print(f'E[X + Y] = {sum(X)/len(X) + sum(Y)/len(Y)}')
```

## Problem 0-3
An increasing subarray of an integer array is any consecutive sequence of array integers whose values strictly increase. Write Python function count-long-subarrays(A),
which accepts Python Tuple A = (a0, a1, . . . , an−1) of n > 0 positive integers, and returns the
number of longest increasing subarrays of A, i.e., the number of increasing subarrays with length at
least as large as every other increasing subarray. For example, if A = (1,3,4,2,7,5,6,9,8),
your program should return 2 since the maximum length of any increasing subarray of A is three
and there are two increasing subarrays with that length: specifically, subarrays (1,3,4) and
(5,6,9).

```python
#1st solution

def count_long_subarrays(A):
    countermax = 1
    counter = 1

    for i in range(len(A)-1):
        
        if A[i+1] > A[i]:
            counter += 1
        elif counter > countermax:
            countermax = counter
            counter = 1
        elif counter <= countermax:
            counter = 1
        

    countermax = max(countermax, counter)
    
    
    counterlength = 0
    counter = 1
    
    for i in range(len(A)-1):
        
        if A[i+1] > A[i]:
            counter += 1
        elif counter == countermax:
            counterlength += 1
        else:
            counter = 1
    return counterlength  
```

```python
count_long_subarrays((1,3,4,2,7,5,6,9,8))
```

```python
#2nd solution

def count_long_subarrays(A):
    
    counter = 1
    max_length = 1
    count_max_length = 1
    
    for i in range(len(A)-1):
        if A[i] < A[i+1]:
            counter += 1
        else:
            counter = 1
        
        if max_length == counter:
            count_max_length += 1
            
        elif counter > max_length:
            max_length = counter
            count_max_length = 1
            
    return count_max_length
```

```python
count_long_subarrays((1,3,4,2,7,5,6,9,8))
```
